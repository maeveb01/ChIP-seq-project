### Peak and Motif detection with GEM

```python
# run GEM without any parameters
!java -jar /usr/local/gem/gem.jar
```

```python
# add parameters that restrict memor and cpu usage
!java -Xmx4G -jar /usr/local/gem/gem.jar --t 1
```

```python
# add miminum set of parameters for a peak detection run
    #use 3gb ram max = Xmx4g
    #use 1 cpu only = --t 1
    #data from chromosome A (chipA.bam) only
    #output file = bam format
!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chipA.bam --f BAM \
    --d /usr/local/gem/Read_Distribution_default.txt
```

```python
# list content of newly created output directory
!ls -lh out
```

```python
#prediction of consensus motif requires additional parameters

!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chipA.bam \
    --out out1 \
    --f BAM \
    --d /usr/local/gem/Read_Distribution_default.txt \
    --genome . \
    --k 10 \
    --g chrom_sizes.txt 
```

```python
#gem run on entire genome

!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chip.bam \
    --out out2 \
    --f BAM \
    --d /usr/local/gem/Read_Distribution_default.txt \
    --genome . \
    --k 10 \
    --g chrom_sizes.txt 
```

```python
#gem using conrtrol file and entire genome 
!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chip.bam \
    --ctrl ctrl.bam \
    --out out3 \
    --f BAM \
    --d /usr/local/gem/Read_Distribution_default.txt \
    --genome . \
    --k 10 \
    --g chrom_sizes.txt 
```

```python
#gem using outBED to load into IGV for chipA
!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chipA.bam \
    --ctrl ctrlA.bam \
    --out out5 \
    --outBED \
    --f BAM \
    --d /usr/local/gem/Read_Distribution_default.txt \
    --genome . \
    --k 10 \
    --g chrom_sizes.txt 
```

```python
#outbed for whole genome
!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chip.bam \
    --ctrl ctrl.bam \
    --out out6 \
    --outBED \
    --f BAM \
    --d /usr/local/gem/Read_Distribution_default.txt \
    --genome . \
    --k 10 \
    --g chrom_sizes.txt 
```

```python
#outbed whole genome with quality control trimming at 40
!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chip_q30.bam \
    --f BAM --genome . \
    --g chrom_sizes.txt \
    --d Read_Distribution_default.txt \
    --k_min 8 --k_max 13 \
    --ctrl ctrl_q30.bam \
    --out out7 \
    --outBED 
```

```python
#outbed whole genome with quality control trimming at 40 and PCR removed
!java -Xmx4G -jar /usr/local/gem/gem.jar \
    --t 1 \
    --expt chip_q30_rmdup.bam \
    --f BAM --genome . \
    --g chrom_sizes.txt \
    --d Read_Distribution_default.txt \
    --k_min 8 --k_max 13 \
    --ctrl ctrl_q30_rmdup.bam \
    --out out8 \
    --outBED 
```
